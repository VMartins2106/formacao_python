# Utilizando List Comprehension

```python
print([x for x in range(800_000, 801_000) if all(x % y != 0 for y in range(2, x))])
```

# Utilizando uma função e o método filter()

```python
def e_primo(numero):
  for i in range(2,numero):
    if (numero % i) == 0:
      return False
  return True
```

```python
primos = list(filter(e_primo, range(2,2000)))
```

```python
print(primos)
```
