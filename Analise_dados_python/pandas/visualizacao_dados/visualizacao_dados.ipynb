## Gráfico de linhas

 ### Matplotlib
 - [Documentação gráficos de linha](https://matplotlib.org/2.0.2/api/lines_api.html)

```python
df.head()
```

```python
df['sepal_length'].plot()
```

```python
df['sepal_length'].plot(
    figsize=(15,5), # TAMANHO DA FIGURA 
    linewidth=(2), # GROSSURA DA LINHA
    color="green", # COR DA LINHA
    title="Tamanho da sépala", # TITULO DO GRÁFICO
    xlabel='índice', # NOME PARA O EIXO X
    ylabel="tamanho em cm", # NOME PARA O EIXO Y
    marker='o', # MARCOS EM TODOS OS PARES ORDENADOS
    linestyle='-.', # ESTILO DA LINHA
    legend=True # LEGENDA DO GRÁFICO
)
```

```python
plt.figure(figsize=(15,5)) # CONFIGURANDO TAMANHO DO GRÁFICO

plt.plot(df.index, df['sepal_length']) # CONFIGURANDO EIXO X E Y
plt.plot(df.index, df['sepal_width'], color='gray',linewidth=1, linestyle='--') # CONFIGURANDO OUTRA LINHA
plt.title('Sépala') # TITULO
plt.xlabel("índice") # NOME EIXO X
plt.ylabel("Tamanho em cm") # NOME EIXO Y
plt.legend(['altura','largura']) # LEGENDA
plt.show() # MOSTRA O GRÁFICO
```

### Subplots

```python
df.head()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA PRIMEIRA COLUNA
plt.plot(df.index, df['sepal_length'], color='green')
plt.title('Altura da sépala')
plt.ylabel('tamanho em cm')

plt.subplot(1,3,2) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA SEGUNDA COLUNA
plt.plot(df.index, df['sepal_width'], color='red')
plt.title('Largura da sépala')
plt.ylabel('tamanho em cm')

plt.subplot(1,3,3) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA TERCEIRA COLUNA
plt.plot(df.index, df['petal_length'],color='yellow')
plt.title('Altura da pétala')
plt.ylabel('tamanho em cm')

plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.subplot([1,1]) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA PRIMEIRA COLUNA
plt.plot(df.index, df['sepal_length'], color='green')
plt.title('Altura da sépala')
plt.ylabel('tamanho em cm')

plt.subplot([1.2]) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA SEGUNDA COLUNA
plt.plot(df.index, df['sepal_width'], color='red')
plt.title('Largura da sépala')
plt.ylabel('tamanho em cm')

plt.subplot([2,1]) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA TERCEIRA COLUNA
plt.plot(df.index, df['petal_length'],color='yellow')
plt.title('Altura da pétala')
plt.ylabel('tamanho em cm')

plt.subplot(2,2,[2,2]) # CONFIGURA UMA LINHA, 3 COLUNAS E ESTÁ NA TERCEIRA COLUNA
plt.plot(df.index, df['petal_length'],color='yellow')
plt.title('Altura da pétala')
plt.ylabel('tamanho em cm')

plt.show()
```

 - [Documentação gráfico de barras horizontais](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)

```python
# utilizando o Matplotlib

plt.figure(figsize=(15,5))
plt.bar(x=df['species'], height=df['species'].count())
plt.title('Contagem de espécies')
plt.xlabel('Espécie')
plt.ylabel('Contagem')
```

```python
# barras horizontais
plt.figure(figsize=(15,5))
plt.barh(width=df['species'].count(), y=df['species'])
plt.title('Contagem de espécies')
plt.xlabel('Contagem')
plt.ylabel('Espécie')
```

### Seaborn (gráfico de barras)

- [Documentação gráfico de barras](http://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
import numpy as np 

df_agrupado =  df.groupby('species').agg({'sepal_length': np.median})
type(df_agrupado)
```

```python
df_agrupado
```

```python
# barras verticais
sns.barplot(data=df_agrupado, x=df_agrupado.index, y='sepal_length')
```

```python
# barras horizontais (TROCAR X PELO Y E Y PELO X)
sns.barplot(data=df_agrupado, x='sepal_length', y=df_agrupado.index)
plt.title('Barras horizontais')
```

## Boxplot (Gráfico / Diagrama de caixa)

- Gráfico muito útil que nos ajuda a identificar:
    - limite inferior
    - limite superior
    - quartis
    - outliers
    - formato da distribuição

### Matplotlib 
- [Documentação boxplot](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html)

```python
df.head()
```

```python
df['sepal_length'].plot(kind='box', figsize=(6, 6))
# RISCOS INFERIOR E SUPERIOR SÃO OS LIMITES INFERIORES E SUPERIORES DE VALORES
# A CAIXA É A MAIOR CONCENTRAÇÃO DE VALORES
# RISCA ACIMA DO MEIO: TERCEIRO QUARTIL,75%
# MEIO: SEGUNDO QUARTIL, MEDIANA, 50%
# RISCA ABAIXO DO MEIO: PRIMEIRO QUARTIL,25%
```

```python
plt.boxplot(df['sepal_length'])
plt.show()
```

### Seaborn
- [Documentação boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
plt.figure(figsize=(4, 4))
sns.boxplot(data=df, x='species', y='sepal_length')
# ESTE PONTO PRETO É UM 'OUTLIER', POIS FOGE MUITO DA MEDIA
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(data=df, x='sepal_length', y='species')
# ESTE PONTO PRETO É UM 'OUTLIER', POIS FOGE MUITO DA MEDIA
```

## Nuvem de palavras (Wordcloud)

- `pip install wordcloud`
- [Documentação](https://github.com/amueller/word_cloud)

```python
%pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

### Lendo o arquivo TXT

```python
with open('wordcloud/elefante.txt', 'r', encoding='utf8') as file:
    texto = file.read()
```

```python
print(texto)
```

```python
wc = WordCloud(height=400, width=400)
```

```python
type(wc)
```

```python
nuvem = wc.generate(texto)
```

### Plotando a nuvem de palavras

```python
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```

### Stopwords

```python
stopwords = ['o', 'a', 'um', 'uma', 'â', 'e', 'ela', 'ele', 'não', 'mas', 'que', 'por', 'ali']
```

```python
wc = WordCloud(height=400, width=400, stopwords=stopwords)
```

```python
nuvem = wc.generate(texto)
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```

### Criando uma máscara



```python
from PIL import Image
import numpy as np
```

```python
imagem = Image.open('wordcloud/elefante.png')
```

```python
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

```python
mascara = np.array(imagem)
```

```python
type(mascara)
```

```python
mascara.shape
```

```python
wc = WordCloud(mask=mascara, 
               background_color='white', 
               stopwords=stopwords,
               contour_color='black',
               contour_width=1)
```

```python
nuvem = wc.generate(texto)
```

```python
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```
