## Agrupando dados

### a função groupby

```python
groupby()
```

````python
df.head()
````

````python
df_agrupado_nome_loja = df.groupby('nome_loja')
````

````python
df_agrupado_nome_loja.median()
````

````python
df.groupby('nome_loja')['lucro'].sum().round(2)
````

### Funções de agregação

````python
df.groupby('descricao_produto').agg({
    'valor_final': np.sum,
    'valor_unitario': np.mean,
    'lucro': np.sum
})

# PARA O VALOR FINAL: ELE SOMA TUDO
# PARA O VALOR UNITÁRIO: ELE CALCULA A MÉDIA
# PARA O LUCRO: ELE SOMA TUDO
````

### Agrupando dados pelo índice

````python
df.groupby(df.index.year).agg({
    'valor_final': np.sum,
    'lucro': np.sum
})
````

### Agrupando dados por mais de uma coluna

````python
df.head()
````

````python
df.groupby(['nome_loja', 'nome_funcionario']).agg({
    'preco_custo': np.sum,
    'lucro': np.sum
})
````

````python
df.head()
````

````python
df.groupby(['categoria', 'descricao_produto']).agg({
    'valor_unitario': np.sum,
    'lucro': np.sum    
})
````

### Agrupando dados por tabelas dinâmicas (Pivot Tables)

```python
pivot_table()
```

````python
df.pivot_table(index='nome_funcionario', aggfunc=np.sum)
````

## Iterando sobre um DataFrame

### iteritems()

- retorna uma tupla
- nome_coluna, conteúdo

````python
df.head()
````

````python
for col_name, col_content in df.iteritems():
    print(col_content)
    break
````

### iterrows()

- retorna uma tupla
- nome da linha, conteúdo da linha

````python
df.head()
````

````python
for row_name, row_content in df.iterrows():
    print(row_name, row_content)
    break
````
